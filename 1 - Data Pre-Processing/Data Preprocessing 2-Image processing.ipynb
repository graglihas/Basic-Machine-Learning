```python
from PIL import Image
from matplotlib import pyplot as plt
import cv2
```

```python
#reading the data
img = cv2.imread('test_image.jpg')
plt.imshow(img)
```

```python
#changing color channels
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
#turning into grayscale
gray = img[:,:,1]
plt.imshow(gray)
```

```python
#applying bilateral filtering
median = cv2.bilateralFilter(gray,5,500,10)
plt.imshow(median)
```

```python
#Applying contrast enhancement
clahe = cv2.createCLAHE(clipLimit=100,tileGridSize=(5,5))
cl1 = clahe.apply(median)
plt.imshow(cl1
        )
```

```python
#Resize image data
cl2 = cv2.resize(cl1, (512,512))
```

```python
ret, th1 = cv2.threshold(cl2, 235, 235, cv2.THRESH_BINARY)
plt.imshow(th1)
```
