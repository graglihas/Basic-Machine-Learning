```python
# importing required python libraries
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
#Assigning column headers to dataset
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
#reading the dataset and assigning column headers
df = pd.read_csv("auto.csv", names = headers)
```

```python
#displayinf top 10 rows of the dataset
df.head(10)
```

```python
#Replacing the missing values with shown with '?' with 'NAN' for ease of computation
df.replace("?",np.nan,inplace =True)
df.head(10)
```

```python
#Checking which is the missing data
missing_data = df.isnull()
missing_data.head(10)
```

```python
#Calculating the mean of normalized-losses column
average_normalized_loss = df["normalized-losses"].astype("float").mean(axis=0)
#Replacing the missing values by mean
df["normalized-losses"].replace(np.nan, average_normalized_loss, inplace=True)
df.head(10)
```

```python
missing_doors = df.loc[df["num-of-doors"].isnull()]
print(missing_doors)
```

```python
#Replacing with mode of data
df["num-of-doors"].replace(np.nan,df["num-of-doors"].value_counts().idxmax(), inplace=True)
```

```python
missing_doors = df.loc[df["make"]=="dodge"]
print(missing_doors)
```

```python
missing_price = df.loc[df["price"].isnull()]
print(missing_price)
```

```python
#dropping the rows
df.dropna(subset=["price"],axis=0,inplace=True)
df.head(10)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
#Data Standardozation
df['city-L/100km'] = 235/df["city-mpg"]
df.head(10)
```

```python
df["length"]
```

```python
#Data Normalization
df['length'] = df['length']/df['length'].max()
df["length"]
```

```python
#Ordinal encoding
encoder = OrdinalEncoder()
df[["drive-wheels"]] = encoder.fit_transform(df[["drive-wheels"]]).astype("int")
df.head(10)
```

```python
#OneHot Encoding
ohe= OneHotEncoder(sparse=False)
result = ohe.fit_transform(df[['body-style']])
result
```

```python
LE = LabelEncoder()
result = LE.fit_transform(df['body-style'])
result
```
