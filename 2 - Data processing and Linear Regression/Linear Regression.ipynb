```python
#Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error as mse
%matplotlib inline
```

```python
#Reading dataset
df=pd.read_csv("automobileEDA.csv")
df.head(10)
```

```python
lr = LinearRegression() #Creating Linear Regression Object
X=df[['engine-size']] #settin independent variable
Y=df['price'] #setting dependent variable
lr.fit(X,Y) #Fitting X,Y for Linear regression
Yhat=lr.predict(X) #predicting a output for X
Yhat
```

```python
print(lr.intercept_) #the value of bias b
print(lr.coef_) #the value of weight w
```

```python
#Calculating Mean squared error
MSE=mse(Y,Yhat)
MSE
```

```python
#Calculating R-squared error
#R=1-(SSE/SST)=SSR/SST
R=lr.score(X,Y)
R
```

```python
#Plotting regression line
plt.figure(figsize=(12,10))
sns.regplot(x="engine-size", y="price", data =df)
plt.ylim(0,)
#df[['x1','x2','x3']].corr()
```

```python
#plotting residual plot
plt.figure(figsize=(12, 10))
sns.residplot(df['engine-size'], df['price'])
plt.show()
#uneven spread
```

```python
#Multiple Linear Regression
X_= df[['horsepower','curb-weight','highway-mpg']] #Setting x1,x2,x3 independent variables
Y_= df['price'] #setting dependent variables
lr.fit(X_,Y_)
Y_hat= lr.predict(X_)
Y_hat
```

```python
print(lr.intercept_) #the value of bias
print(lr.coef_) #weights w1,w2,w3
```

```python
#Mean Squared error
MSE_ = mse(Y_,Y_hat)
MSE_
```

```python
#R-squared error
R_=lr.score(X_,Y_)
R_
```

```python
plt.figure(figsize=(12,10))
ax1 = sns.distplot(Y_, hist=False, color="r", label="Actual Value") #plotting Actual values
sns.distplot(Y_hat, hist=False, color="b", label="Fitted Values" , ax=ax1) #plotting predixted values
plt.title('Actual vs Fitted Values for Price')
plt.xlabel('Price (in dollars)')
plt.show()
```
