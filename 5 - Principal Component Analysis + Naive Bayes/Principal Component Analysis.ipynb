```python
#Importing modules and libraries
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
#Reading the dataframe
df=pd.read_csv("FuelConsumptionCo2_modified.csv")
df.head(10)
```

```python
#Setting Independent and Dependent Variable
X=df[['MODELYEAR','ENGINESIZE','CYLINDERS','FUELCONSUMPTION_CITY','FUELCONSUMPTION_HWY','FUELCONSUMPTION_COMB','FUELCONSUMPTION_COMB_MPG']]
Y=df['CO2EMISSIONS']
```

```python
lr = LinearRegression()
lr.fit(X,Y)
R=lr.score(X,Y)
R
```

```python
#Standardize data
X_std = StandardScaler().fit_transform(X)
X_std[:5]
```

```python
#Calculating Coovariance matrix
cov_matrix = np.cov(X_std,rowvar=False)
cov_matrix[:5]
```

```python
#Calculate Eigen decomposition
e_value, e_vector = np.linalg.eig(cov_matrix)
```

```python
e_value[:5]
```

```python
e_vector[:5]
```

```python
#Percentage of knowledge in features
info_percent = (e_value/np.sum(e_value))*100
info_percent
```

```python
projected_1 = X_std.dot(e_vector.T[0])
projected_2 = X_std.dot(e_vector.T[1])

res = pd.DataFrame(projected_1, columns=['PC1'])
res['PC2'] = projected_2
res['Y'] = Y
res.head()
```

```python
lr_ = LinearRegression()
X_ = res[['PC1','PC2']]
lr.fit(X_,Y)
R_=lr.score(X_,Y)
R_
```
