```python
#Importing the modules and libraries
import numpy as np
import matplotlib.image as img
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA, IncrementalPCA
```

```python
#Raading Image
img_org = img.imread("OIP.jpg")
plt.figure(figsize=[12,8])
plt.imshow(img_org)
```

```python
#Standardize image data
img_org = img_org/255
```

```python
#Calculating component percentage
img_sum = img_org.sum(axis=2)
img_sum_=img_sum.reshape((img_sum.shape[0], img_sum.shape[1], 1))
img_sum_ = img_sum_ + 0.000000001
img_ratio = img_org/img_sum_
```

```python
#Applying PCA
pca = PCA(n_components=0.85)
pca.fit(img_sum)
components=pca.transform(img_sum)
projected =pca.inverse_transform(components)
```

```python
#Displaying image
img_sum_=projected.reshape((projected.shape[0], projected.shape[1], 1))
img_final = img_ratio*img_sum_
plt.figure(figsize=[12,8])
plt.imshow(img_final)
```
